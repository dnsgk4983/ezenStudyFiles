```python
# 4칙연산 계산기 클래스 구현
# 생성자가 없는 경우
class FourCal:
    # 두 수의 입력값을 받는 하위함수
    def setdata(self, first, second):
        self.first = first
        self.second = second

    # 입력받은 두 수를 self로 내려받아 더하는 하위 함수를 정의합니다.
    def add(self):
        result = self.first + self.second
        return result

    # 입력받은 두 수를 self로 받아서 곱하는 하위함수입니다.
    def mul(self):
        result = self.first * self.second
        return result
    
    # 두 수를 빼는 하위함수입니다.
    def sub(self):
        result = self.first - self.second
        return result
    
    # 두 수를 나누는 하위함수입니다.
    def div(self):
        result = self.first / self.second
        return result
```

```python
# 생성자가 존재하는 클래스 예시
class FourCal:
    # 두 수의 입력값을 받는 하위함수 (생성자 버전)
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result

    def mul(self):
        result = self.first * self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
```

```python
# 생성자가 있으므로 클래스 괄혼안에 인풋 데이터를 넣어야한다.
f = FourCal(2, 7)
g = FourCal(5, 55)

f.div()
g.mul()
```

```python
# '클래스의 상속
# 상속을 받는 하위 클래스가 상위 부모클래스의 모든 기능을 사용할 수 있다.
# 코드의 재사용 + 확장성
# 상속받는 클래스 = 자식 클래스 // 상속해 주는 클래스 = 부모 클래스 입니다.
# 상속받는 클래스는 괄호안에 반드시 부모 클래스 이름을 써야한다~~~

# 클래스 상속 예제 (위 FourCal 클래스 상속)
class MoreMath(FourCal):
    # 자식 클래스에서 div 함수는 이제 몫을 구한다.
    def div(self):
        print("상속 일어나서 부모클래스의 함수와는 다릅니다.")
# 클래스의 상속이 된 것을 확인할 수 있다.
# 부모클래스의 생성자가 기능을 받아서 99, 20 입력값을 처리한 것을 확인!!
# 부모 클래스의 하위 함수인 add를 상속받아서 자식 클래스가 사용하는 것을 확인!
a = MoreMath( 99, 20)
a.add()
#자식클래스의 div함수가 부모클래스의 div 함수를 덮어씀
# 클래스 안의 함수 이름이 같으면 자식클래스 함수에 우선권이 있다.
a.div()
```

```python
# 클래스 상속 응용 예제
# 부모클래스의 함수를 통해서 개발 로직 테스트 가능

class FourCal:
    # 두 수의 입력값을 받는 하위함수 (생성자 버전)
    def __init__(self, first, second):
        self.first = first
        self.second = second
    # 입력받은 두 수를 self로 내려받아 더하는 하위 함수를 정의합니다.
    def add(self):
        print("add 함수 아직 안만드셨네요")

class MoreMath(FourCal):
    # 자식 클래스에서 div 함수는 이제 목슬 구한다.
    def mul(self):
        print("add 함수 개발이 끝났습니다.")

# 자식 클래스 소속 add 함수를 안만들면
# 부모클래스 add함수가 실행 되면서 개발이 되지 않았다고 알려줌
ppp = MoreMath(7, 77)
ppp.add()
```

```python
# BeautifulSoup
# BeautifulSoup은 파이썬에서 HTML 및 XML 문서를 파싱하고 분서하는데 사용되는 라이브러리입니다.
# 이를 통해 웹 스크래핑 및 데이터 추출 작업을 수행 할 수 있습니다.
# 아래는 BeautifulSoup 라이브러리의 기본 사용법에 대한 개요입니다.

# 파싱 = 데이터의 패턴을 분석해서 데이터의 구조를 해석하는 행위
# --> 데이터의 패턴을 파악하는 것이 크롤링 핵심!!!!
```

```python
# ***** 파싱:
# BeatuifulSoup 객체를 생성하려면 HTML 또는 XML 문자열과
# 파서를 전달해야 합니다. 일반적으로 'html.parser'를 사용합니다.

# 사이트 url을 변수에 담아서 선언
# 사이트 주소를 요청하는 라이브러리

# 1. 라이브러리 호출
import urllib.request as req
from bs4 import BeautifulSoup

# 2. 원하는 url 주소 담아서 열리는지 확인
url = "https://media.daum.net/"
# 위 주소를 열어봐서 잘 열리는 경우에만 정보 열람
res = req.urlopen(url)
source = res.read()

# 3. 한글깨짐 방지
# 한글을 지원하는 유니코드입니다. (한글깨짐 방지)
source = source.decode('utf-8')

# 4. 데이터 해석 (parsing)
# BeautifulSoup 에서 본 사이트는 html 문서라는 것을 알려줌
html = BeautifulSoup(source, 'html.parser')
# ***파악한 패턴을 적용하여 원하는 정보를 atags 변수에 저장 합니다.
atags = html.select('a[class=link_txt]')
# print(atags)
# print(len(atags))

# 뉴스 내용만 수집하려면?
# result_data 변수에 빈 리스트 자료구조를 선언 (뉴스 헤드라인 데이터만 담을라고)
result_data = []

cnt = 0 # 총 뉴스 헤드라인의 개수도 세고싶어서 cnt 변수 선언
# 크롤링한 전체 데이터에서 개별 데이터 호출
# for문으로 크롤링 데이터 셋에서 개별문장 하나씩 호출
for ii in atags:
    cnt += 1
    # 데이터를 문자열로 변환 ( 전체 문장에서 문자열 데이터만 추출함 )
    atag_str = str(ii.string)
    # 걸러진 문자열 데이터를 준비한 리스트 자료구조에 담습니다.
    result_data.append(atag_str.strip()) # 문장좌우에 불필요한 여백은 strip으로 제거

print(result_data)
```
